```python
def mandelbrot_cell(x):
    '''
    Funkcja sprawdza czy liczba x jest w zbiorze 
    Mandelbrota z pewną dokładnością

    Argumenty:
    x (krotka) - liczba zespolona
    '''
    p = complex(x[0], x[1])
    z = complex(0)
    for i in range(100):
        z = z**2 + p
        if abs(z) > 2:
            return False
    return True
```

```python
import numpy as np


def mandelbrot(N=500):
    '''
    Funkcja rysuje zbiór Mandelbrota.

    Argumenty:
    N (int) - dokładność obrazka (domyślnie 500)
    '''
    X = np.linspace(-2, 2, N)
    wynik = np.array([mandelbrot_cell([i, j]) for j in X for i in X])
    return wynik.reshape(N, N)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.matshow(mandelbrot())
```

```python
def julia_cell(x, y):
    '''
    Funkcja sprawdza czy liczba x jest w zbiorze 
    Julii z pewną dokładnością

    Argumenty:
    x (krotka) - liczba zespolona
    '''
    z = complex(x[0], x[1])
    c = complex(y[0], y[1])
    for i in range(100):
        z = z**2 + c
        if abs(z) > 2:
            return False
    return True
```

```python
def julia(c, N=500):
    '''
    Funkcja rysuje zbiór Julii

    Argumenty:
    c (int) - liczba zespolona
    N (int) - dokładność obrazka (domyślnie 500)
    '''
    X = np.linspace(-2, 2, N)
    wynik = np.array([julia_cell([i, j], c) for j in X for i in -X])
    return wynik.reshape(N, N)
```

```python
plt.matshow(julia([-0.1, 0.65]))
```

```python
plt.matshow(julia([0.285, 0.015]))
```
